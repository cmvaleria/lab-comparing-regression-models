### 9. Create a function that given a model prediction and real values returns a pandas dataframe with the following table:

```python
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error, mean_absolute_percentage_error 
from math import sqrt
    
```

```python
def evaluation_dframe(y,yp):
    data = {'R2 ':[r2_score(y, yp)],
            'MAE':[mean_absolute_error(y, yp)],
            'MSE':[mean_squared_error(y,yp)],
            'RMSE':[sqrt(mean_squared_error(y,yp))], 
            'MAPE':[mean_absolute_percentage_error(y,yp)] }
    df = pd.DataFrame(data).T
    return df

evaluation_dframe(y_test,y_pred_test).rename(columns={0: 'test_evaluation'})
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
```

```python
models = [LinearRegression(),KNeighborsRegressor(n_neighbors=2),MLPRegressor(random_state=1, max_iter=500)]
def train_models(models, X_train, y_train):
    trained_models = []
    r2score = []
    MSE = []
    MAPE = []
    for model in models:
        model.fit(X_train, y_train)
        y_pred_test = model.predict(X_test_new_scaled_df)
        trained_models.append(model)
        r2score.append(r2_score(y_test,y_pred_test))
        MSE.append(mean_squared_error(y_test,y_pred_test))
        MAPE.append(mean_absolute_percentage_error(y_test,y_pred_test))
        
    print('R2', trained_models, r2score)
    print('MSE', trained_models, MSE)
    print('MAPE', trained_models, MSE)
```
